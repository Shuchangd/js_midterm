```python
import nltk
import string
nltk.download('stopwords');
punctuation = list(string.punctuation)
stop_word = list(set(nltk.corpus.stopwords.words('english')))
stopwords = stop_word + punctuation
for idx, lists in enumerate(xiamen['formatted_text']):
    all_words = [word for word in lists if word not in stopwords]
    xiamen['formatted_text'][idx] = all_words
xiamen.head()
```
